```python
def make_match_results_dict(match):
    match_results=defaultdict(list)   
    if match["teams"][0]["win"]=="Win":
        match_results["team_100_win"].append(True)
    else:
        match_results["team_100_win"].append(False)

    for champion in match["participants"]:
        if champion["teamId"]==100:
            match_results["team_100"].append(champion["championId"])
        else:
            match_results["team_200"].append(champion["championId"])
    return match_results
```
