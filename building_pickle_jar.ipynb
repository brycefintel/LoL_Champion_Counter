### Build synergy and matchup dicts.  Remove entries with less than 40/50 entries.

```python
synergy = pickle.load( open( "picklejar/backup/synergy_win_percent2.p", "rb" ) )
matchup = pickle.load( open( "picklejar/backup/matchup_win_percent2.p", "rb" ) )

```

```python
sorted(list(matchup.items()),reverse=True,key= lambda x: x[1])
```

```python
count=0
for key in matchup.keys():
    if matchup[key][1]<50:
        matchup[key]=(.5,matchup[key][1])
        count+=1
count
```

```python
count=0
for key in synergy.keys():
    if synergy[key][1]<40:
        synergy[key]=(.5,matchup[key][1])
        count+=1
count
```

```python
pickle.dump( synergy, open( "picklejar/trimmed_synergy.p", "wb" ) )
pickle.dump( matchup, open( "picklejar/trimmed_matchup.p", "wb" ) )
```

### Build lane dict.  Remove entries with less than 30 entries.

```python
lane_total = pickle.load( open( "picklejar/backup/total_lane.p", "rb" ) )
lane_win = pickle.load( open( "picklejar/backup/wins_lane.p", "rb" ) )

```

```python
lane={}
for key in lane_total.keys():
    lane[key]=(lane_win[key]/float(lane_total[key]),lane_total[key])
```

```python
sorted(list(lane.items()),reverse=True,key= lambda x: x[1])
```

```python
count=0
for key in lane.keys():
    if lane[key][1]<30:
        lane[key]=(.5,lane[key][1])
        count+=1
count
```

```python
for key in matchup.keys():
    if key not in lane.keys():
        if key[0] != key [1]:
            print key
            lane[key]=.5
```

```python
pickle.dump( lane, open( "picklejar/trimmed_lane.p", "wb" ) )
```

### Build reverse champ_dict

```python
champion_dict = pickle.load(open("picklejar/champion_dict.p","rb"))
```

```python
reverse_champ_dict={}
for x in champion_dict["data"].keys():
    strip=(str(champion_dict["data"][x]["name"]).replace(" ","").replace("'","").replace(".","").lower())
    reverse_champ_dict[strip]=int(x)
```

```python
pickle.dump( reverse_champ_dict, open( "picklejar/reverse_champ_dict.p", "wb" ) )
```
